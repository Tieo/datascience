```
def get_url(item_name, code_df): 
    code = code_df.query("name=='{}'".format(item_name))['code'].to_string(index=False) 
    print(code[1:])
    url = 'http://finance.naver.com/item/sise_day.nhn?code={code}'.format(code=code[1:]) 
    print("요청 URL = {}".format(url)) 
    return url

sss_name = sss['name']
s_name = list(sss_name)

def get_season_price(company, code_df):
  item_name= company
  url = get_url(item_name, code_df) 
  #url

  df = pd.DataFrame() # 1페이지에서 10페이지의 데이터만 가져오기 
  for page in range(1, 11): 
      pg_url = '{url}&page={page}'.format(url=url, page=page) 
      df = df.append(pd.read_html(pg_url, header=0)[0], ignore_index=True)

  df = df.dropna()
  code_df = code_df.dropna()
  df = df[['날짜','종가']] #날짜와 종가만 가져옴


  season10 = df.iloc[:11]# 5월 8일부터 6월 12일까지 약 한달
  season9 = df.iloc[11:21]
  season8 = df.iloc[21:31]
  season7 = df.iloc[31:41]
  season6 = df.iloc[41:51]
  season5 = df.iloc[51:61]
  season4 = df.iloc[61:71]# 5월 8일부터 6월 12일까지 약 한달
  season3 = df.iloc[71:81]
  season2 = df.iloc[81:91]
  season1 = df.iloc[91:]
  s10_m = season4.mean()# 5월간 종가 평균
  s9_m = season3.mean()
  s8_m = season2.mean()
  s7_m = season1.mean()
  s6_m = season4.mean()# 5월간 종가 평균
  s5_m = season3.mean()
  s4_m = season4.mean()# 5월간 종가 평균
  s3_m = season3.mean()
  s2_m = season2.mean()
  s1_m = season1.mean()
  s_m = pd.concat([s1_m, s2_m, s3_m, s4_m,s5_m, s6_m, s7_m, s8_m,s9_m, s10_m], axis =1)# keys= ['Season1','Season2','Season3' ,'Season4'] , axis = 1) # 좌우로 합침
  #type(s4_m)
  s_m = [item_name, s1_m[0], s2_m[0], s3_m[0], s4_m[0],s5_m[0], s6_m[0], s7_m[0], s8_m[0],s9_m[0], s10_m[0]] # type: list 
  return s_m

season_price = []
for i in s_name:
  company_prices = get_season_price(i, code_df)
  season_price.append(company_prices)

season_price 
```
